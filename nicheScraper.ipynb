```python
#import libraries
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd
import random
```

```python
#get list of top 100 cities for young adults
if __name__ == "__main__":
    options = uc.ChromeOptions()
    options.headless = False
    driver = uc.Chrome(
        use_subprocess=False,
        options=options,
    )
    cities = []
    driver.get("https://www.niche.com/places-to-live/search/best-cities/?whoLivesHere=youngProfessionals")

    for page_num in range(0, 4):  # Loop through the first 4 pages

        # Find and scrape all target elements on the current page
        try:
            elements = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, "//h2[@class='MuiTypography-root MuiTypography-labelMedium MuiLink-root MuiLink-underlineHover nss-i5xnri' and @aria-hidden='true' and @data-testid='search-result__title']"))
            )
            if elements:
                for element in elements:
                    cities.append(element.text)
            else:
                print(f"No matching elements found on page {page_num}.")
        except Exception as e:
            print(f"An error occurred on page {page_num}: {e}")

        # If it's page 4, we stop here
        if page_num == 4:
            break

        # Try to click the "Next" button to move to the next page
        try:
            next_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Go to next page']"))
            )
            next_button.click()
            time.sleep(random.uniform(2, 5))  # Add a slight random delay to allow the next page to load
        except Exception as e:
            print(f"Could not navigate to page {page_num + 1}: {e}")
            break

    # Close the browser
    driver.quit()
```
