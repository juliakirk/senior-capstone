```python

url = 'https://api.yelp.com/v3/businesses/search'

# Authorization headers
headers = {
    'Authorization': f'Bearer {key}'
}

locations = pd.read_csv('../datasets/data.csv')

results = []

for index, row in locations.iterrows():
    latitude = row['Latitude']
    longitude = row['Longitude']

    params = {
        'term': 'bars',     
        'latitude': latitude,
        'longitude': longitude,
    }
    print(params)

    response = requests.get(url, headers=headers, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        total_businesses = data.get('total', 0)

        # Append the result to the list
        results.append({
            'latitude': latitude,
            'longitude': longitude,
            'total_bars': total_businesses
        })
        
        print(f"Location ({latitude}, {longitude}) - Total delis: {total_businesses}")
    else:
        print(f"Error for location ({latitude}, {longitude}): {response.status_code}")

    # To avoid hitting rate limits, you might need to add a delay between requests
    time.sleep(1)  # Sleep for 1 second to be safe

# Convert the results into a DataFrame and save it to a new CSV
results_df = pd.DataFrame(results)
results_df.to_csv('yelp_results.csv', index=False)

print("Results saved to 'yelp_results.csv'.")
```
