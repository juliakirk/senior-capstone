```python
urban_areas_df = pd.read_csv('../datasets/capstoneData.csv', dtype={'CensusCode': str})
codes = urban_areas_df.iloc[:, 3]
codes
```

```python
populations = []
for i in codes:
    response = requests.get(f"https://api.census.gov/data/2023/acs/acsse?get=NAME,K200101_001E&for=urban%20area:{i}&key={key}")
    
    if response.status_code == 200:  # Check if the request was successful
        data = response.json()
        value = data[1][1]
        populations.append(value) 
    else:
        print(f"Error: {response.status_code} for urban area code {i}")
        populations.append(None)

# Add the population data as a new column in the DataFrame
urban_areas_df['Population'] = populations

# Save the updated DataFrame back to a CSV file
urban_areas_df.to_csv('../datasets/capstoneData.csv', index=False)

# Optionally, check the DataFrame to see the new column
print(urban_areas_df.head())
```
